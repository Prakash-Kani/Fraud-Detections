```python
def missing_values(df):
    dic = {'ColumnName':[], 'MissingValues':[], 'Percentage':[]}
    for col in df.columns:
        dic['ColumnName'].append(col)
        null = df[col].isna().sum()
        dic['MissingValues'].append(null)
        dic['Percentage'].append((null/df.shape[0])*100)
    return pd.DataFrame(dic).sort_values('Percentage', ascending= False)
missing_values(df)
```

```python
# missing values are present less than 1%, So that, we will removed missing value rows
df.dropna(axis=0, inplace=True)
missing_values(df)
```

# Preprocessing

```python
df.phone_number[:5]
```

```python
def phone(data):

    if '+' in data or data.count('-')==3:
        data = ''.join(data.split('-')[1:])

    elif '-' in data or '(' in data or '.' in data:
        data = data.replace('-', '')
        data = data.replace('.', '')
        data = data.replace('(', '')
        data = data.replace(')', '')

    else:
        pass
        
    if len(data)==10:
        return data
    else:
        data = data.split('x')[0]
        
        if len(data)==10:
            return data
            
        else:
            print('not valid mobile number')
    # return data
df['phone_number'] =df.phone_number.apply(lambda x: phone(x))
df['phone_number']
```

### User Agent Column

```python
df.user_agent[:10]
```

```python
def browser(data):
    return data.split('/')[0]
df['browser'] =df.user_agent.apply(browser)
```

```python
def operating_system(data):
    if '(' in data:
        d = data
        data = data[data.index('(')+1:data.index(')')]
        data =data.split(';')[0]

        if 'X11' ==data or 'Linux' in data:
            return 'Linux'
        elif 'Windows' in data or 'compatible' in data:
            return 'Windows'
        elif 'Android' in data:
            return 'Android'
        elif 'Macintosh' in data:
            return 'Macintosh'
        elif 'iPhone' in data:
            return 'iPhone'
        elif 'iPod' in data:
            return 'iPod'
        elif 'iPad' in data:
            return 'iPad'
            
        else:
            pass

df['operating_system'] =df.user_agent.apply(operating_system)
```

```python
def search_engine(data):
    if ')' in data:
        d = data
        data = data[data.index(')')+1:]
        
        if ''==data:
            # print(data,d)
            return 'Trident'
        else:
            return data.split('/')[0]
         
    
df['search_engine'] =df.user_agent.apply(search_engine)
```

```python
def security_level(data):
    if '(' in data:
        d = data
        data = data[data.index('(')+1:data.index(')')]
        # data =data.split(';')[0]
        if ' U;' in data:
            return 1
        else:
            return 0

df['security_level'] =df.user_agent.apply(security_level)
```

```python
# Removing user_agent column
df.drop('user_agent', axis =1, inplace = True)
```
